```python
import requests
from bs4 import BeautifulSoup

headers = {'Accept': 'image/avif,image/webp,image/apng,image/svg+xml,image/*,*/*;q=0.8',
          'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/106.0.0.0 Safari/537.36'}

top_films = []
genres_film = []
for i in range(1,6):
    url = f'https://ru.kinorium.com/collections/imdb/233/?order=sequence&page={i}&perpage=50&show_viewed=1'
    source = requests.get(url, headers=headers)
    source.encoding = 'utf-8'

    soup = BeautifulSoup(source.text, 'lxml')
    titles = soup.find_all(class_='movie-title__text filmList__item-title-link-popup link-info-movie-type-film')
    genres = soup.find_all(class_="filmList__extra-info")
    
    top_films.extend([i.text.strip().replace('\xa0', ' ') for i in titles])
    genres_film.extend([i.text.strip().replace('\xa0', ' ').split(',')[0] for i in genres])

# film_genre = {x:y for x,y in zip(top_films, genres_film)}
# print(len(film_genre),'\n', film_genre)

with open('films.txt', 'w') as amogus:
    for i in zip(top_films, genres_film):
        amogus.write(i[0] + ': ' + i[1] +'\n')
```

```python
import pandas as pd

data = pd.DataFrame(columns=['title', 'genre'])
data['title'] = top_films
data['genre'] = genres_film
data = data.set_index('title')
data
```
